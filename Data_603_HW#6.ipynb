### Question2:
Write a Spark program that reads your browser history file, then displays the top 5 websites you visited in the last week?

```python
sc.stop()
#import required modules
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
 
# Create spark context and sparksession
sc = SparkContext("local", "Spark read url")
spark = SparkSession(sc)

#Convert the json file into a dataframe
df1 = spark.read.option("multiline","true").json("C:\\Users\\isanj\\Downloads\\history.json")
df2 = df1["url", "visitCount"]

#Convert the dataframe to rdd
rdd2 = df2.rdd
# rdd2.take(5)

#Count the number of time I visited a website

urlCounts = rdd2.reduceByKey(lambda a,b: a + b)
# print(urlCounts.collect())

df3 = sc.parallelize(urlCounts.collect()).toDF(["url", 'count'])
df3_sorted = df3.sort(desc("count"))
df3_sorted.show(5)
```

```python
print("The websites I visted the most in the past week are:", "\n")

for row in df3_sorted.take(5):
    print(row['url'], ",", row['count'], "times.")
```

```python
sc.stop()
sc = SparkContext("local", "read file")

rdd = sc.textFile('C:\\Users\\isanj\\OneDrive\\Desktop\\Data#603 Summer 2020\\Applied_Data_Science_HW#6.txt')
```

```python
rdd.take(5)
```

```python
 rdd1 = rdd.map(lambda line: line.lower())

rdd2 = (rdd1.flatMap(lambda line: line.split(" "))).map(lambda word: (word, 1))
#print(rdd2.collect())
wordCounts = rdd2.reduceByKey(lambda a,b: a + b)
```

```python

output = wordCounts.collect()
for (word, count) in output:
    print("%s: %i" % (word, count))
    
```

```python
from pyspark import SparkContext, SparkConf, sql

from pyspark.sql import Row

sc = SparkContext.getOrCreate()

sqlContext = sql.SQLContext(sc)

df = sc.parallelize(output).toDF(["word", 'count'])
df.show(10)
```

```python

df_lenght = df.filter(length("word")>5)
#df_lenght.show()
df_sorted = df_lenght.sort(desc("count"))
df_sorted.show(10)
```

```python
df_top_five = df_sorted.take(5)
print(df_top_five)
```

```python
print("the most used words in the Applied Data Science textbook are:")

for row in df_top_five:
    print("<<",row['word'],">>", "occurred", "<<", row['count'], ">>", "times.")

```
